# Theoretical Answers

# 1. What is NumPy?
NumPy is a Python library used for numerical computing. It supports large, multi-dimensional arrays and matrices, and provides mathematical functions to operate on them efficiently.

# 2. How does broadcasting work in NumPy?
Broadcasting allows NumPy to perform operations on arrays of different shapes by 'stretching' the smaller shape to match the larger one.

# 3. What is a Pandas DataFrame?
A DataFrame is a 2-dimensional labeled data structure in Pandas that allows you to store and manipulate tabular data easily.

# 4. Explain the use of the groupby() method in Pandas
groupby() groups data using some criteria and allows you to apply functions like sum(), mean(), etc., on each group.

# 5. Why is Seaborn preferred for statistical visualizations?
Seaborn simplifies the creation of complex statistical plots with built-in themes, colors, and aggregated data support.

# 6. Differences between NumPy arrays and Python lists
NumPy arrays are faster, require less memory, and support element-wise operations unlike Python lists.

# 7. What is a heatmap, and when should it be used?
A heatmap displays data values using color and is useful for visualizing correlation matrices or intensity values.

# 8. What does “vectorized operation” mean in NumPy?
Vectorized operations perform computations directly on entire arrays, without using explicit Python loops.

# 9. How does Matplotlib differ from Plotly?
Matplotlib produces static plots suitable for publications, while Plotly provides interactive, web-based visualizations.

# 10. Significance of hierarchical indexing in Pandas?
"""It allows multiple index levels, enabling better organization and subsetting of complex datasets."""

# 11. Role of Seaborn’s pairplot() function?
pairplot() creates pairwise scatterplots of numerical columns to show relationships between variables.

# 12. Purpose of the describe() function in Pandas?
describe() gives summary statistics (mean, std, min, max, etc.) for each column in a DataFrame.

# 13. Why is handling missing data important in Pandas?
Missing data can lead to incorrect analysis. Pandas provides methods like dropna(), fillna() to handle it.

# 14. Benefits of using Plotly for visualization?
Plotly offers interactive, publication-ready, browser-based plots including 3D charts.

# 15. How does NumPy handle multidimensional arrays?
NumPy uses ndarray objects that support operations like slicing, reshaping, and broadcasting across multiple dimensions.

# 16. Role of Bokeh in data visualization?
Bokeh creates interactive plots and dashboards that can be viewed in web browsers.

# 17. Difference between apply() and map() in Pandas?
map() works element-wise on a Series, while apply() works on both Series and DataFrames with custom functions.

# 18. Advanced features of NumPy?
Includes linear algebra, broadcasting, masked arrays, FFT, random sampling, etc.

# 19. How does Pandas simplify time series analysis?
Pandas supports datetime indexing, resampling, shifting, and rolling window operations out-of-the-box.

# 20. Role of pivot table in Pandas?
Pivot tables allow summarization and aggregation of data similar to Excel pivot tables.

# 21. Why is NumPy’s array slicing faster than list slicing?
NumPy arrays use contiguous memory and are optimized in C, making slicing faster than Python lists.

# 22. Common use cases for Seaborn?
Seaborn is used for correlation plots, distribution visualization, regression plots, and categorical data analysis.


```python
# Practical Coding Part

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# 1. Create 2D NumPy array and calculate row sum
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Row sums:", np.sum(arr, axis=1))
```

```python
# 2. Find mean of a column in Pandas
df = pd.DataFrame({'A': [10, 20, 30], 'B': [5, 15, 25]})
print("Mean of column A:", df['A'].mean())
```

```python
# 3. Scatter plot using Matplotlib
plt.scatter(df['A'], df['B'])
plt.title("Scatter Plot of A vs B")
plt.xlabel("A")
plt.ylabel("B")
plt.show()
```

```python
# 4. Correlation matrix and heatmap
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 5. Bar plot using Plotly
fig = px.bar(df, x='A', y='B', title='Bar Plot of A vs B')
fig.show()
```

```python
# 6. Add new column based on existing
df['C'] = df['A'] + df['B']
print("New DataFrame with column C:\n", df)
```

```python
# 7. Element-wise multiplication of two arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("Element-wise multiplication:", a * b)
```

```python
# 8. Line plot with multiple lines
plt.plot([1, 2, 3], [4, 5, 6], label='Line 1')
plt.plot([1, 2, 3], [6, 5, 4], label='Line 2')
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

```python
# 9. Filter rows where column A > 15
filtered_df = df[df['A'] > 15]
print("Filtered rows where A > 15:\n", filtered_df)
```

```python
# 10. Histogram using Seaborn
sns.histplot(df['A'], kde=True)
plt.title("Histogram of A")
plt.show()

```

```python
# 11. Matrix multiplication using NumPy
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])
result = np.dot(m1, m2)
print("Matrix Multiplication Result:\n", result)
```
